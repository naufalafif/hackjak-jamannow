```python
# Import
import pandas as pd
from openpyxl import load_workbook
from IPython.display import Image
from sklearn.preprocessing import Imputer
from sklearn import tree
from subprocess import check_call
import re
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score
import graphviz 
```

```python
wb = load_workbook(filename = 'data.xlsx')
sheet_ranges = wb.active
df = pd.DataFrame(sheet_ranges.values)
df.columns = ['NO_RM','UMUR','JK','TAHUN','JENIS_PENYAKIT','GEJALA']
df = df[1:][['UMUR','JK','JENIS_PENYAKIT','GEJALA']]
data = df[['UMUR','JK']]

data['UMUR'].replace(['1-9 Thn.'], 0,inplace= True)
data['UMUR'].replace(['10-19 Thn.'], 1,inplace= True)
data['UMUR'].replace(['20 - 49 Thn.'], 2,inplace= True)
data['UMUR'].replace(['20-49 Thn.'], 2,inplace= True)
data['UMUR'].replace(['50-69 Thn.'], 3,inplace= True)
data['UMUR'].replace(['>70 Thn.'], 4,inplace= True)
data['JK'].replace(['L'], 0,inplace= True)
data['JK'].replace(['P'], 1,inplace= True)
```

```python
list_gejala = df.GEJALA.tolist()
myLists = [i.split(',') for i in list_gejala] 
all_list = []

for myList in myLists:
    for y in myList:
        y = y.replace(" ", "")
        all_list.append(y)

columns_gejala = list(set(all_list))

for column in columns_gejala:
        data[column] = 0

for index, row in data.iterrows():
    for x in myLists[index-1]:
        x = x.replace(" ", "")
        if x in columns_gejala:
            data.loc[index, x] = 1
    
data
```

```python
y = df['JENIS_PENYAKIT'].values
features = data
features_columns = ['UMUR','JK'] + columns_gejala
# print features_columns
imp = Imputer(axis=0)
x = imp.fit_transform(features)
clf = tree.DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(x,y)
```

```python
with open("tree_baru.dot", "w") as f:
    f = tree.export_graphviz(clf, out_file=f, feature_names=features_columns,
                            filled=True, rounded=True,
                            special_characters=True)

from subprocess import check_call
check_call(['dot','-Tpng','tree_baru.dot','-o','OutputFile.png'])
# dot_data = tree.export_graphviz(clf, out_file=None, feature_names=features_columns,
#                             filled=True, rounded=True,
        
#                                 special_characters=True)
# graph = graphviz.Source(dot_data)
# graph
# graph.draw('file.png')
```
